```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.pandas.set_option('display.max_columns',None)
```

```python
dt = pd.read_csv('data/train.csv')
print(dt.shape)
```

```python
dt.head()
```

# Data Analysis
1 missing values
2 numerical values
3 distr of nv
4 categorical values
5 cardinility of cat va
6 outliers
7 relationship b/w dependent and independent values

```python
features_with_na = [features for features in dt.columns if dt[features].isnull().sum()>1]
for feature in features_with_na:
    print(feature,np.round(dt[feature].isnull().mean(),4),' % missing values')
```

# let check relationship of missing values with dependent  

```python
for feature in features_with_na:
    data = dt.copy()
    # let make a variable where the values is 1 when the observation is true else 0
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

## finding numerical variables

```python
numerical_features = [feature for feature in dt.columns if dt[feature].dtype != "O"]
print('length of numerical features  = ',len(numerical_features))
dt[numerical_features].head()
```

## temporal Variables

```python
year_features = [feature for feature in numerical_features if 'Yr' in  feature or 'Year' in feature]
year_features
```

```python
#let explore the feature of this year 
```

```python
for feature in year_features:
    print(feature,dt[feature].unique())
```

```python
# let analyze these temporal variables
```

```python
dt.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year Sold")
plt.ylabel('Median House Price')
plt.title('House Price Vs Year')
```

```python
for feature in year_features:
    if feature != 'YrSold':
        data = dt.copy()
        data[feature] = data['YrSold'] - data[feature] 
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        
        plt.show()
        
```

```python
# Know getting discrete and continuous feature
discrete_features = [feature for feature in numerical_features if len(dt[feature].unique()) < 25 and feature not in year_features+['Id']]
print('discrete variables = ',len(discrete_features))
```

```python
dt[discrete_features].head()
```

```python
# let find relationship b/w sales price and discrete features
for feature in discrete_features:
    data = dt.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
```

```python
# finding continous features
continous_features = [feature for feature in numerical_features if feature not in discrete_features+year_features+['Id']]
print('Continous Features : ',len(continous_features))
```

```python

# let check the distribution of continous data
for feature in continous_features:
    data = dt.copy()
    data[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

# Exploratory Data Analysis part #2

```python
# we will using logrithmic function
for feature in continous_features:
    data = dt.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]  = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

## Outliers

```python
for feature in continous_features:
    data = dt.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]  = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

## categorical Feature

```python
categorical_feature = [feature for feature in dt.columns if dt[feature].dtypes == 'O']
print('Categorical Features : ',len(categorical_feature))
```

```python
for feature in categorical_feature:
    print(f'the Feature is {feature}  and number of Categ {len(dt[feature].unique())}')
```

```python
# finding relationship b/w categorical feature and Dependent Features
```

```python
for feature in categorical_feature:
    data = dt.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
    
```
